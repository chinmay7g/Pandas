```python
import pandas as pd
nba = pd.read_csv(r'C:\Users\Administrator\Desktop\Python Test Data\nba.csv')
```

```python
nba.head()
```

```python
teammates = nba.groupby('Team')
```

```python
teammates.first()
```

```python
#accessing the information related to a particular group

teammates.get_group('Los Angeles Lakers')
```

```python
#so we learnt that
# on applying the first groupby command, we actually create an instance(object) of the class(something on similar lines)
#and then we apply get_group to that same object
```
