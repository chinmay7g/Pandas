```python
#getting started with .loc method
```

```python
#we can use this method to subset a single row as a series

#this method gives us a comprehensive information regarding all column labels included in the dataframe

data.loc['Kobe Bryant']
```

```python
#choosing multiple rows

data.loc[['Kobe Bryant', 'Jae Crowder']]
```

```python
#choosing via slicing
data.loc['Avery Bradley' : 'Jordan Mickey']
```

```python
#selecting rows and columns

row = ['Kobe Bryant', 'Avery Bradley', 'Amir Johnson']
column = ['Team', 'Number', 'Weight','College']

data.loc[row, column]
```

```python
#selecting some rows and all the columns

data.loc[['Kobe Bryant', 'Avery Bradley', 'Amir Johnson'], :]
```

```python
#getting on with iloc method

#selecting and giving a row as a series
data.iloc[3]
```

```python
#selecting a position by a factor of 2
data.iloc[5::20]
```

```python
#selecting columns and rows

data.iloc[3:6, [1,4]]
```
